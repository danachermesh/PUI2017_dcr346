# SPLIT BY CATEGORY

Formatting the dataframe to better fit to the assignment qustions.

### (1) trip duration day vs night:

## H0: there is no statistical difference between tripduration at nighttime and tripduration in daytime
$$ \alpha = 0.05 $$

### (2) age of bikers trip's origin Manhattan vs Brooklyn:

## H0: there is no statistical difference between age of bikers which their trip was originated in Manhattan and in Brooklyn
$$ \alpha = 0.05 $$

```python
# (1) adding DayOrNight column

# creating a function to define daytime / nighttime trip
def day_or_night(hour):
    if hour >= 7 and hour < 20:
        return 1
    else:
        return 0

df['DayOrNight'] = df['hour'].apply(day_or_night)
df = df.dropna(0)
df.head()
```

```python
# (2) adding ageM (Male) and ageF (Female) columns

df['ageM'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 1)]
df['ageF'] = 2015 - df['birth year'][(df['usertype'] == 'Subscriber') & (df['gender'] == 2)]

df['ageM'].dropna(inplace= True)
df['ageF'].dropna(inplace= True)
df.head()
```

```python
# (2) assigning lat-lon to describe Manhattan boundaries

minlat = 40.705987
maxlat = 40.807772
maxlon = -73.935288
minlon = -74.018589

def findLatLong(lat_lon):
    lat, lon = lat_lon
    if ((lat > minlat) & (lat < maxlat) & (lon > minlon) & (lon < maxlon)):
        return 1
    else:
        return 0
    
df['Manhattan'] = df[['start station latitude','start station longitude']].apply(findLatLong,axis=1)
df['Manhattan'].dropna(inplace = True)
```

```python
df["Age"]= 2015 - df['birth year']
df['Age'].dropna(inplace = True)
df.head()
```

## KS tests to compare 2 samples

http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.ks_2samp.html

the KS test in scipy returns the p-value. the Null hypothesis will be rejected if:

$D_KS(n1,n2) > c(\alpha) \sqrt{\frac{(n1 + n2)}{n1n2}}$

```python
# converting daytime trip durations and night time trip durations from dataframe to np.arrays
daytrips = (df.tripduration[df.DayOrNight == 1]).values
nighttrips = (df.tripduration[df.DayOrNight == 0]).values

# converting age & trip origin from dataframe to np.arrays
manhattan = df.Age[df.Manhattan == 1].values
brooklyn = df.Age[df.Manhattan == 0].values
```

```python
ks_trip = scipy.stats.ks_2samp(daytrips, nighttrips)
ks_origin = scipy.stats.ks_2samp(manhattan, brooklyn)

print("ks_trip:", ks_trip)
print("ks_origin:", ks_origin)
```

#### KS test results:

In both cases ((1) trip duration and (2) trip origin) the KS result is ~0.05. Referring to the table where ($\alpha$) = 0.05, c($\alpha$) should be > 1.36 to reject the Null hypothesis. Since the pvalue I got is much smaller (very close to 0), it is much further at the tail of the distribution and **I reject the (1),(2) Null hypotheses**.

# Pearson's  test for correlation

    
http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.pearsonr.html#scipy.stats.pearsonr



```python
#sorting the np.arrays
daytrips.sort()
nighttrips.sort()
manhattan.sort()
brooklyn.sort()

#randomly picking daytrips/manhattan to set each pair of arrays' lenght to be the same
daytimerandom = np.random.choice(daytrips, len(nighttrips), replace=False)
manrandom = np.random.choice(manhattan,len(brooklyn), replace=False)
# Pearson's test
pears_trip = scipy.stats.pearsonr(daytimerandom, nighttrips)
pears_origin = scipy.stats.pearsonr(manrandom, brooklyn)

print("pears_trip:", pears_trip)
print("pears_origin:", pears_origin)
```

#### Pearson's test results:

**I reject the (1),(2) Null hypotheses**.

# Spearman's  test for correlation

http://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.spearmanr.html#scipy.stats.spearmanr

```python
spear_trip = scipy.stats.spearmanr(daytimerandom, nighttrips)
spear_origin = scipy.stats.spearmanr(manrandom, brooklyn)

print("spear_trip:",spear_trip)
print("spear_origin:",spear_origin)
```

#### Spearman's test results:

**I reject the (1),(2) Null hypotheses**.
