### Assignment 1: Write an ipython notebook that demonstrates visually in a data-driven way the Central Limit Theorem. 

- GENERATE  100 samples of different sizes N (N>10 & N<2000) from each of 5 different distributions (500 samples in total), _all with the same population mean_
- Include a _Normal_, a _Poisson_, a _Binomial_, a _Chi-Squared_ distribution, and 1 more of your choice.    

```python
import pandas as pd
import numpy as np 
import pylab as pl
import seaborn

%pylab inline
```

```python
np.random.seed(999)

mean = 77

samples_norm = []
samples_poi = []
samples_bin = []
samples_chi = []
samples_gam = []

for i in range(100):
    samples_norm.append(np.random.normal(mean,1,size=np.random.randint(10,2000)))
    samples_poi.append(np.random.poisson(mean,size=np.random.randint(10,2000)))
    samples_bin.append(np.random.binomial(2*mean,.5,size=np.random.randint(10,2000)))
    samples_chi.append(np.random.chisquare(mean,size=np.random.randint(10,2000)))
    samples_gam.append(np.random.gamma(mean,1,size=np.random.randint(10,2000)))
```

- For each sample plot the sample mean (dependent var.) against the sample size N (independent var.) (if you want you can do it with the sample standard deviation as well). 

```python
fig = plt.figure(figsize=(10, 25))
#creating the grid of all charts
grid = plt.GridSpec(5, 1, hspace=0.5)
#orgenizing charts
ax0 = plt.subplot(grid[0, :2])
ax1 = plt.subplot(grid[1, :2])
ax2 = plt.subplot(grid[2, :2])
ax3 = plt.subplot(grid[3, :2])
ax4 = plt.subplot(grid[4, :2])

for sample in range(len(samples_norm)):
    ax0.plot(len(samples_norm[sample]), np.mean(samples_norm[sample]), 'o')
for sample in range(len(samples_poi)):
    ax1.plot(len(samples_poi[sample]), np.mean(samples_poi[sample]), 'o')
for sample in range(len(samples_bin)):
    ax2.plot(len(samples_bin[sample]), np.mean(samples_bin[sample]), 'o')
for sample in range(len(samples_chi)):
    ax3.plot(len(samples_chi[sample]), np.mean(samples_chi[sample]), 'o')
for sample in range(len(samples_gam)):
    ax4.plot(len(samples_gam[sample]), np.mean(samples_gam[sample]), 'o')
    
#lables and titles
ax0.set_title("Normal distribution", fontsize = 16)
ax1.set_title("Poisson distribution", fontsize = 16)
ax2.set_title("Binomial distribution", fontsize = 16)
ax3.set_title("Chi-Squared distribution", fontsize = 16)
ax4.set_title("Gamma distribution", fontsize = 16)
for ax in (ax0, ax1, ax2, ax3, ax4):
    ax.set_xlabel("Size of Sample", fontsize = 12)
    ax.set_ylabel("Mean of Sample", fontsize = 12)
```
